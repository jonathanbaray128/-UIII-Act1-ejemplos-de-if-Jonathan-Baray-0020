Ejemplo de if

```python
# Ejemplo 1: Estructura if simple con Operadores Relacionales
# Este código verifica si una persona ha alcanzado la mayoría de edad
# .
edad = 20 # Se asigna el valor entero 20 a la variable edad [14]
if edad >= 18: # Evalúa si el valor de edad es mayor o igual a 18 (operador relacional) [10, 13]
    print("Eres mayor de edad.") # Imprime el mensaje solo si la condición anterior es verdadera [13]
```

```python
# Ejemplo 2: Estructura if-else con Operadores Aritméticos
# Este ejemplo determina si un número es par o impar utilizando el operador de módulo
# .
numero = 7 # Se define la variable numero con el valor 7 [14]
if numero % 2 == 0: # El operador % obtiene el resto de la división; == 0 verifica si es par [7, 15]
    print("El número es par.") # Se ejecuta si el resto de la división entre 2 es cero [15]
else: # Bloque que se ejecuta si la condición del if resulta falsa [2]
    print("El número es impar.") # Imprime el mensaje indicando que el número no es par [15]
```

```python
# Ejemplo 3: Estructura if-elif-else con Operadores Lógicos
# Este ejemplo clasifica la temperatura de un sistema utilizando rangos lógicos combinados
# .
temp = 25 # Define la variable temp con el valor 25 [14]
if temp < 0: # Verifica si la temperatura es menor a cero (congelación) [10]
    print("Clima gélido") # Se ejecuta si la condición temp < 0 es verdadera
elif temp >= 0 and temp <= 20: # Usa 'and' para verificar que la temperatura esté en un rango cerrado [12, 16]
    print("Clima frío o templado") # Se ejecuta si ambas comparaciones relacionales son verdaderas [12]
elif temp > 20 and temp <= 30: # Evalúa si la temperatura está en un segundo rango específico [2]
    print("Clima agradable") # Se ejecuta si el valor está entre 21 y 30 inclusive [16]
else: # Caso por defecto si ninguna de las condiciones anteriores se cumplió [2]
    print("Clima caluroso") # Mensaje final para temperaturas mayores a 30 [17]
```

```python
# Ejemplo 4: Condición Compleja (Año Bisiesto)
# Este ejemplo combina operadores aritméticos, relacionales y lógicos en una sola evaluación compleja
# .
anio = 2024 # Asigna el año a evaluar a la variable anio [13]
# Verifica si es divisible por 4 y NO por 100, O si es divisible por 400 (usa %, ==, !=, and, or) [9, 11, 13]
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.") # Imprime el resultado si la lógica booleana es verdadera [13]
else: # Si el año no cumple las reglas matemáticas de bisiesto [2]
    print(f"{anio} no es un año bisiesto.") # Imprime el mensaje de negación [13]
```
