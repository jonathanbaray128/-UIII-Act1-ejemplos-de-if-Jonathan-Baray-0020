```python
# ==========================================
# INVESTIGACIÓN POO: CLASE CONTADOR
# ==========================================

# Definición de la Clase
class Contador:
    def __init__(self, nombre, telefono, edad):
        # Atributos: Datos referentes al contador
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Declarar Impuestos
    def declarar_impuestos(self):
        print(f"Resultado: El contador {self.nombre} ha enviado la declaración fiscal con éxito.")

    # Función 2: Auditar Cuentas
    def auditar_cuentas(self):
        print(f"Resultado: {self.nombre} terminó de revisar los libros. ¡Las cuentas cuadran perfectamente!")

    # Función 3: Generar Nómina
    def generar_nomina(self):
        print(f"Resultado: La nómina de la empresa ha sido calculada y procesada por {self.nombre}.")

# --- Creación de Objetos (Instancias) ---
# Nota: Siguiendo tu solicitud, los nombres de las variables son contador1 y contador2

# Objeto 1
contador1 = Contador("Lic. Roberto Ruiz", "555-123-456", 38)

# Objeto 2
contador2 = Contador("Lic. Ana Martínez", "555-987-654", 29)

# --- Ejecución de Funciones y Mensajes de Salida ---

print("--- REPORTE DEL CONTADOR 1 ---")
print(f"Nombre: {contador1.nombre}")
print(f"Teléfono: {contador1.telefono}")
print(f"Edad: {contador1.edad}")
# Llamada a las acciones (funciones)
contador1.declarar_impuestos()
contador1.generar_nomina()

print("\n" + "-"*30 + "\n")

print("--- REPORTE DEL CONTADOR 2 ---")
print(f"Nombre: {contador2.nombre}")
print(f"Teléfono: {contador2.telefono}")
print(f"Edad: {contador2.edad}")
# Llamada a las acciones (funciones)
contador2.auditar_cuentas()
contador2.declarar_impuestos()

print("Programa terminado")
print("Programa realizado por Jonathan Baray NL = 06")
```
